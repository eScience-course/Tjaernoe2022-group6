## Chlorophyll trends for polar region

```python
min_lat= 60
max_lat= 90
min_lon= 0
max_lon= 360
var='chlos'
cel_type='areacello'
model= 'NorESM2-LM'

files_dir='/escience2022/Ada/monthly/chlos_Omon_NorESM2-LM_historical_r1i1p1f1_gn*.nc'

inp_dms=[files_dir,model,min_lat,max_lat,min_lon,max_lon,var,cel_type]

chlos=md.regional_average(inp_dms)

y=chlos.to_numpy()
x=chlos.year

xx=x.to_numpy()
chlos=[xx,y]

a=np.array(chlos)
np.savetxt('Data/timeseries/chlos.txt',a)
```
